```python
import pandas as pd
```

```python
df = pd.DataFrame({'col1':{'row1':1, 'row2':2, 'row3':3}, 'col2': {'row1':4, 'row3':5, 'row4':6}, 'col3' : {'row1':3, 'row2':5, 'row6': 10, 'row7': 7}})
print(df)
print("----------------")
```

```python
print(df.count()) #default to axis=0, column counts
print(df.count(axis=1))
print(df.min())
print(df.max(axis=1))
print("----------------")
print(df.describe())
```

```python
n_most = df.nlargest(2,['col2'])
print(n_most)
```
